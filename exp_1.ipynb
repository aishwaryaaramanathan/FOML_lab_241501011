```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Iris dataset
df = sns.load_dataset('iris')

# Univariate Plots

# Sepal Width
df_Setosa = df.loc[df['species'] == 'setosa']
df_Virginica = df.loc[df['species'] == 'virginica']
df_Versicolor = df.loc[df['species'] == 'versicolor']

plt.scatter(df_Setosa['sepal_width'], np.zeros_like(df_Setosa['sepal_width']) + 0, label='Setosa')
plt.scatter(df_Virginica['sepal_width'], np.zeros_like(df_Virginica['sepal_width']) + 1, label='Virginica')
plt.scatter(df_Versicolor['sepal_width'], np.zeros_like(df_Versicolor['sepal_width']) + 2, label='Versicolor')
plt.xlabel('sepal.width')
plt.legend()
plt.show()

# Sepal Length
plt.scatter(df_Setosa['sepal_length'], np.zeros_like(df_Setosa['sepal_length']) + 0, label='Setosa')
plt.scatter(df_Virginica['sepal_length'], np.zeros_like(df_Virginica['sepal_length']) + 1, label='Virginica')
plt.scatter(df_Versicolor['sepal_length'], np.zeros_like(df_Versicolor['sepal_length']) + 2, label='Versicolor')
plt.xlabel('sepal.length')
plt.legend()
plt.show()

# Petal Width
plt.scatter(df_Setosa['petal_width'], np.zeros_like(df_Setosa['petal_width']) + 0, label='Setosa')
plt.scatter(df_Virginica['petal_width'], np.zeros_like(df_Virginica['petal_width']) + 1, label='Virginica')
plt.scatter(df_Versicolor['petal_width'], np.zeros_like(df_Versicolor['petal_width']) + 2, label='Versicolor')
plt.xlabel('petal.width')
plt.legend()
plt.show()

# Petal Length
plt.scatter(df_Setosa['petal_length'], np.zeros_like(df_Setosa['petal_length']) + 0, label='Setosa')
plt.scatter(df_Virginica['petal_length'], np.zeros_like(df_Virginica['petal_length']) + 1, label='Virginica')
plt.scatter(df_Versicolor['petal_length'], np.zeros_like(df_Versicolor['petal_length']) + 2, label='Versicolor')
plt.xlabel('petal.length')
plt.legend()
plt.show()

# Bivariate Plots

# Sepal Width vs Petal Width
sns.FacetGrid(df, hue='species', height=5).map(sns.scatterplot, "sepal_width", "petal_width").add_legend()
plt.show()

# Sepal Length vs Petal Length
sns.FacetGrid(df, hue='species', height=5).map(sns.scatterplot, "sepal_length", "petal_length").add_legend()
plt.show()

# Multivariate Plot (Pairplot)
sns.pairplot(df, hue="species", height=2)
plt.show()

```
